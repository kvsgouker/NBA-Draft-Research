```python

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36"
}

# download: A good citizen download function
#     url - the url accessed
#     destination - local file to write
#
# respects code 429 and waits instead of pounding.

def download(url, destination):
    # filename = prefix + "-" + os.path.basename(url)
    if not os.path.exists(destination):
        try:
            # Send a GET request with headers
            response = requests.get(url, headers=headers)
            # Check if the request was successful
            if response.status_code == 200:
                with open(destination, 'w') as f:
                    f.write(response.text)
                print("Downloaded " + destination)
            elif response.status_code == 429:
                # Extract the Retry-After header value
                retry_after = response.headers.get("Retry-After")
                if retry_after:
                    # Convert the Retry-After value to seconds
                    retry_after_seconds = int(retry_after)
                    print("Rate limit exceeded. Waiting for " + str(retry_after_seconds) + " seconds.")
                    time.sleep(retry_after_seconds)
                    # Retry the request after waiting
                    download(url, destination)
                else:
                    print("Rate limit exceeded. Retry-After header not found.")
            else:
                print("Website returned " + str(response.status_code))
        except urllib.error.HTTPError:
            print("Failed to download " + url)
        except Exception:
            print("Error writing " + destination)
    return

```

## Compute the annual draft power of a team given its list of picks.

```python
# getAverageWinShare: determines average win share for a given draft rank
def getAverageWinShare(df, rank):
    draft_history = df[df["Rank"] == rank]
    win_share = draft_history["WS"].item()
    return win_share / (FINAL_YEAR - FIRST_YEAR)

def getPickPower(df, rank):
    # Positions 1-5: Calculate the average win share at that draft ranking
    if rank <= 5:
        draft_power = getAverageWinShare(df, rank)
    # Positions 6-55: Blur the values by taking the mean of nearby draft rankings
    elif rank <= 55:
        blurred_values = [getAverageWinShare(df, i) for i in range(rank-2, rank+3)]
        draft_power = np.mean(blurred_values)
    # Positions 56-60: Calculate the average win share of positions 55-60
    else:
        blurred_values = [getAverageWinShare(df, i) for i in range(55, 60)]
        draft_power = np.mean(blurred_values)
    return draft_power

# subset_drafted_df
def getDraftPower(df, draft_rankings):
    total_draft_power = 0
    # Iterate over each draft ranking
    for rank in draft_rankings:
        total_draft_power += getPickPower(df, rank)
    return total_draft_power

# Prints a title decorated by stars.
def printFancyTitle(title):
    # Calculate the length of the title
    title_length = len(title)

    # Print the title with decoration
    print("*" * (title_length + 4))
    print(f"* {title} *")
    print("*" * (title_length + 4))

# Format all stats the same
def printFormattedTestStat(value, dec=6):
    format_string = "{:."+str(dec)+"f}"
    return format_string.format(value)
```

```python

def GoMining(df, dependent_variable):
    variables = []
    for name in df.columns:
        try:
            if df[name].var() < 1e-7:
                continue

            formula = dependent_variable + ' ~ ' + name
            formula = formula.encode('ascii')

            model = smf.ols(formula, data=df)
            if model.nobs < len(df)/2:
                continue

            results = model.fit()
        except (ValueError, TypeError):
            continue
        except patsy.PatsyError:
            raise ValueError(MESSAGE)

        variables.append((results.rsquared, name))

    return variables

def MiningReport(variables, n=30):
    """Prints variables with the highest R^2.

    t: list of (R^2, variable name) pairs
    n: number of pairs to print
    """
    all_vars = ReadVariables()

    variables.sort(reverse=True)
    for mse, name in variables[:n]:
        key = re.sub('_r$', '', name)
        try:
            desc = all_vars.loc[key].desc
            if isinstance(desc, pandas.Series):
                desc = desc[0]
            print(name, mse, desc)
        except KeyError:
            print(name, mse)

def SummarizeResults(results):
    """Prints the most important parts of linear regression results:

    results: RegressionResults object
    """
    for name, param in results.params.items():
        pvalue = results.pvalues[name]
        print('%s   %0.3g   (%.3g)' % (name, param, pvalue))

    try:
        print('R^2 %.4g' % results.rsquared)
        ys = results.model.endog
        print('Std(ys) %.4g' % ys.std())
        print('Std(res) %.4g' % results.resid.std())
    except AttributeError:
        print('R^2 %.4g' % results.prsquared)
```

## Playoff Records

This section builds a dataframe, all_playoff_df, containing the team statistics for each postseason, which is used to determine the annual success of the team. Specifically, the wins and losses of each playoff round are read from the tables.

```python
if generate_data:
    years = range(FIRST_YEAR, FINAL_YEAR+1)

    if do_playoffs:
        print("\nProcessing Team Playoffs\n")
        for year in years:
            url = FILE_PATH_TO_PLAYOFFS + str(year) + ".html"
            try:
                with open(url, 'r') as f:
                    html = f.read()

                soup = BeautifulSoup(html, 'html.parser')

                # find the table element that contains the playoff data
                table = soup.find("table", id="all_playoffs")

                # create a pandas dataframe from the table
                playoffs_df = pd.read_html(str(table))[0].fillna("")

                # Index past the head. Get all the lines for the year.
                playoff_rows = playoffs_df.iloc[:, 1]
                # Isolate the ones with the word "over" (very inflexible, but we don't control their html anyhow. an api would be better.)
                playoff_rows = playoff_rows[playoff_rows.str.contains('over')]

                # define an empty list to store the extracted data
                playoff_data = []
                row_count = 0
                series_level = 0

                # Loop through each playoff row
                for playoff_row in playoff_rows:
                    match = re.match(r'^(.*?)\s+over\s+(.*?)\s+\((\d+)-(\d+)\)$', playoff_row)
                    if match:
                        field1 = match.group(1).strip()
                        field2 = match.group(2).strip()
                        field3 = int(match.group(3))
                        field4 = int(match.group(4))
                        row_count += 1

                        # Calculate the series level based on row count
                        series_level = int(math.log(row_count, 2)) + 1

                        # Append the extracted data as a tuple to the playoff_data list
                        playoff_data.append((series_level, field1, field2, field3, field4))
                    else:
                        print("Parsing Error: Playoff Team Extraction Error\n")

                # Create a DataFrame from the playoff_data list
                playoff_df = pd.DataFrame(playoff_data, columns=['Series', 'Team 1', 'Team 2', 'Team 1 Wins', 'Team 2 Wins'])

                playoff_df["Year"] = year
                if year==FIRST_YEAR:
                    all_playoff_df = playoff_df
                else:
                    all_playoff_df = pd.concat([all_playoff_df, playoff_df])
                    
            except FileNotFoundError:
                print("File not found: " + url)

            # # print output
            # print("\nYear: "+str(year)+"\n")
            # # print the playoff results
            # print(playoff_df.to_string(index=False))

        # write results to scratch files.
        if write_data:
            all_playoff_df.to_csv(FILE_PATH_TO_COMPILED_DATA + 'all_playoff.csv', index=False)
            all_playoff_df.to_excel(FILE_PATH_TO_COMPILED_DATA + 'all_playoff.xlsx', index=False)
            all_playoff_df.to_json(FILE_PATH_TO_COMPILED_DATA + 'all_playoff.json', orient='records')
else:
    all_playoff_df = pd.read_csv(FILE_PATH_TO_COMPILED_DATA + 'all_playoff.csv')
    
print("All done with playoff records!")
```

Six data frames are created to detail player performance using basketball reference as its statistic source:

```python

def extract_player_info(soup):
    info_div = soup.find('div', {'id': 'info', 'class': 'players'})

    player_name = info_div.find('h1').find('span').text.strip()

    info_dict = {}

    for p in info_div.find_all('p'):
        tag = None
        for child in p.children:
            if tag:
                if tag == 'College:':
                    college_tag = child.find_next('a')
                    college_text = college_tag.text.strip() if college_tag else None
                    info_dict['College'] = college_text
                elif tag == 'Colleges:':
                    college_child = child
                    colleges = []
                    while college_child:
                        if college_child.name == 'a':
                            colleges.append(college_child.string)
                        college_child = college_child.next_sibling
                    info_dict['College'] = colleges 
                elif tag == 'Draft:':
                    draft_team = None
                    rank = None
                    draft_year = None
                    draft_child = child
                    while draft_child:
                        if draft_child.name == 'a':
                            if "NBA Draft" in draft_child.string:
                                draft_year = draft_child.string[:4]  # Extract first 4 characters
                                info_dict['Draft_Year'] = draft_year
                            elif not draft_team:
                                draft_team = draft_child.text.strip()
                                info_dict['Draft_Team'] = draft_team
                        elif "overall" in draft_child.string.lower():
                            rank_match = re.search(r'(\d+)\D*overall', draft_child.string)  # Extract number using regex
                            rank = rank_match.group(1) if rank_match else None
                            info_dict['Rank'] = rank
                        draft_child = draft_child.next_sibling
                elif tag == 'Position:':
                    position = child.strip().split('\n', 1)[0]  # Trim and extract the first line
                    info_dict[tag] = position
                else:
                    info_dict[tag] = child.text.strip()
                tag = None

            if child.name == 'strong':
                tag = child.text.strip()
            elif child.name == 'span':
                if child.text:
                    if re.match(r'\d+-\d+', child.text):
                        height = child.text.strip()
                        info_dict['Height'] = height
                    elif child.text.endswith('lb'):
                        weight = child.text.strip()
                        info_dict['Weight'] = weight[:-2]
                
    # Extract birthdate and birth year
    birth_span = soup.find('span', {'id': 'necro-birth'})
    if birth_span:
        # birth_date = birth_span.find('a').text.strip()
        birth_year = birth_span['data-birth'][:4]
        info_dict['Birthdate'] = birth_span['data-birth']

    player_info = {
        'Player': player_name,
        'Position': info_dict.get('Position:'),
        'Shoots': info_dict.get('Shoots:'),
        'College': info_dict.get('College'),
        'Yrs': info_dict.get('Career Length:'),
        'Team': info_dict.get('Draft_Team'),
        'Rank': info_dict.get('Rank'),
        'Year': info_dict.get('Draft_Year'),
        'Height': info_dict.get('Height'),
        'Weight': info_dict.get('Weight'),
        'Birthdate': info_dict.get('Birthdate')
    }

    return player_info
```

## Add seasonal sum of WS and MP for players who played for team.

* You cannot run this without dropping the tables in the initial block or there will be duplicate column name errors.

```python
# Connect to the database
conn = sqlite3.connect(FILE_PATH_TO_COMPILED_DATA+'nba_data.db')
cursor = conn.cursor()

# Add the 'WS_Total' column to the 'combined_season' table
cursor.execute("ALTER TABLE combined_season ADD COLUMN WS_Total REAL")

# Calculate the sum of 'WS' for each team and year combination and update 'WS_Total' column
cursor.execute("""
    UPDATE combined_season
    SET WS_Total = (
        SELECT SUM(WS)
        FROM advanced_stats
        WHERE combined_season.Tm = advanced_stats.Tm
            AND combined_season.Year = advanced_stats.Year
    )
""")

# Add the 'MP_Total' column to the 'combined_season' table
cursor.execute("ALTER TABLE combined_season ADD COLUMN MP_Total REAL")

# Calculate the sum of 'MP' for each team and year combination and update 'MP_Total' column
cursor.execute("""
    UPDATE combined_season
    SET MP_Total = (
        SELECT SUM(MP)
        FROM advanced_stats
        WHERE combined_season.Tm = advanced_stats.Tm
            AND combined_season.Year = advanced_stats.Year
    )
""")

# Add the 'WS_Playoff_Total' column to the 'combined_season' table
cursor.execute("ALTER TABLE combined_season ADD COLUMN WS_Playoff_Total REAL")

# Calculate the sum of 'WS' for each team and year combination and update 'WS_Playoff_Total' column
cursor.execute("""
    UPDATE combined_season
    SET WS_Playoff_Total = COALESCE((
        SELECT SUM(WS)
        FROM playoff_advanced_stats
        WHERE combined_season.Tm = playoff_advanced_stats.Tm
            AND combined_season.Year = playoff_advanced_stats.Year
    ), 0)
""")


# Add the 'MP_Playoff_Total' column to the 'combined_season' table
cursor.execute("ALTER TABLE combined_season ADD COLUMN MP_Playoff_Total REAL")

# Calculate the sum of 'MP' for each team and year combination and update 'MP_Playoff_Total' column
cursor.execute("""
    UPDATE combined_season
    SET MP_Playoff_Total = COALESCE((
        SELECT SUM(MP)
        FROM playoff_advanced_stats
        WHERE combined_season.Tm = playoff_advanced_stats.Tm
            AND combined_season.Year = playoff_advanced_stats.Year
    ), 0)
""")
    

# Add the 'MP_Playoff_Total' column to the 'combined_season' table
cursor.execute("ALTER TABLE combined_season ADD COLUMN WS_PER_48 REAL")
cursor.execute("ALTER TABLE combined_season ADD COLUMN Playoff_WS_PER_48 REAL")
cursor.execute("ALTER TABLE combined_season ADD COLUMN estimated_season_performance REAL")

# Calculate the new fields: WS_PER_48, PLAYOFF_WS_PER_48, estimated_season_performance
cursor.execute("""
UPDATE combined_season
SET WS_PER_48 = 48 * WS_Total / MP_Total,
    Playoff_WS_PER_48 = CASE
        WHEN MP_Playoff_Total = 0 THEN 0
        ELSE 48 * WS_Playoff_Total / MP_Playoff_Total
    END,
    estimated_season_performance = WS_Total + WS_Playoff_Total * 4
    """)


# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

## Table for Draft Rank Player Performance

```python
# Connect to SQLite database
conn = sqlite3.connect(FILE_PATH_TO_COMPILED_DATA+'nba_data.db')

# Read tables into DataFrames
player_info_df = pd.read_sql_query('SELECT * FROM player_info WHERE Rank <= 60', conn)
advanced_stats_df = pd.read_sql_query('SELECT player_id, MP, OWS, DWS, WS, WS/48 FROM advanced_stats', conn)
playoff_advanced_stats_df = pd.read_sql_query('SELECT player_id, MP, OWS, DWS, WS, WS/48 FROM playoff_advanced_stats', conn)

# Merge the player_info table with the advanced_stats table
merged_df = pd.merge(player_info_df[['player_id', 'Player', 'Rank', 'Year']], advanced_stats_df[['player_id', 'MP', 'WS', 'OWS', 'DWS']], on='player_id')
# Group by player_id, Player, Rank, and Year, summing WS, DWS, and OWS separately
ws_regular_df = merged_df.groupby(['player_id', 'Player', 'Rank', 'Year']).agg({'MP': 'sum', 'WS': 'sum', 'DWS': 'sum', 'OWS': 'sum'}).reset_index()
ws_regular_df.rename(columns={'MP': 'MP_Regular', 'WS': 'WS_Regular', 'DWS': 'DWS_Regular', 'OWS': 'OWS_Regular'}, inplace=True)

# Format columns to three decimal places
ws_regular_df['WS_Regular'] = ws_regular_df['WS_Regular'].round(3)
ws_regular_df['DWS_Regular'] = ws_regular_df['DWS_Regular'].round(3)
ws_regular_df['OWS_Regular'] = ws_regular_df['OWS_Regular'].round(3)

# Merge with playoff_advanced_stats and group by player_id and Name again, summing WS, OWS, and DWS values
merged_df = pd.merge(player_info_df[['player_id', 'Player', 'Rank', 'Year']], playoff_advanced_stats_df[['player_id', 'MP', 'WS', 'OWS', 'DWS']], on='player_id', how='left')
# Group by player_id, Player, Rank, and Year, summing WS, DWS, and OWS separately
ws_playoffs_df = merged_df.groupby(['player_id', 'Player', 'Rank', 'Year']).agg({'MP': 'sum', 'WS': 'sum', 'DWS': 'sum', 'OWS': 'sum'}).reset_index()
ws_playoffs_df.rename(columns={'MP': 'MP_Playoffs', 'WS': 'WS_Playoffs', 'DWS': 'DWS_Playoffs', 'OWS': 'OWS_Playoffs'}, inplace=True)

# Format columns to three decimal places
ws_playoffs_df['WS_Playoffs'] = ws_playoffs_df['WS_Playoffs'].round(3)
ws_playoffs_df['DWS_Playoffs'] = ws_playoffs_df['DWS_Playoffs'].round(3)
ws_playoffs_df['OWS_Playoffs'] = ws_playoffs_df['OWS_Playoffs'].round(3)

# Merge both dataframes to create the final result
result_df = pd.merge(ws_regular_df, ws_playoffs_df, on=['player_id', 'Player', 'Rank', 'Year'], how='left')

# Calculate WS/48 for RS and PO
result_df['WS/48_Regular'] = 48*result_df['WS_Regular']/result_df['MP_Regular']
result_df['WS/48_Playoffs'] = 48*result_df['WS_Playoffs']/result_df['MP_Playoffs']
result_df['WS/48_Regular'] = result_df['WS/48_Regular'].round(3)
result_df['WS/48_Playoffs'] = result_df['WS/48_Playoffs'].round(3)

# just turn nan values to 0 for now
result_df.fillna(0, inplace=True)

preferred_order = ['Player', 'Rank', 'Year', 
                   'MP_Regular', 'WS_Regular', 
                   'WS/48_Regular', 
                   'MP_Playoffs', 'WS_Playoffs', 
                   'WS/48_Playoffs']

result_df = result_df[preferred_order]

# Print the resulting DataFrame using tabulate
columns = ['Name', 'Rank', 'Year Drafted', 'MP RS', 'WS RS', 'WS/48 RS', 'MP PO', 'WS PO', 'WS/48 PO']
print(tabulate(result_df, headers=columns, tablefmt='psql', showindex='never'))

# Build an intermediate csv file with data from data base. We can use it in modeling without all the rebuilding steps.
result_df.to_csv(FILE_PATH_TO_COMPILED_DATA+"career_ws.csv", index=False)
```

## Add players who were drafted and never played.

The previous data considers only the players who were drafted and actually played in the NBA, so successful players who are drafted later are overrepresented. Now I will add dummy records for all draft positions to adjust for players drafted who never played.

"R-squared minus rho-squared" refers to the difference between the coefficient of determination (R-squared) and the square of the correlation coefficient (rho).

R-squared (coefficient of determination): R-squared measures the proportion of the variance in the dependent variable that is explained by the independent variable(s). It ranges from 0 to 1, where a value of 1 indicates that all the variability in the dependent variable is accounted for by the independent variable(s). R-squared can be interpreted as the goodness-of-fit of a regression model. In our case only 14% of the variance can be explained by rank.

rho (correlation coefficient): rho is the correlation coefficient, which measures the strength and direction of the linear relationship between two variables. It ranges from -1 to +1, where -1 indicates a perfect negative linear relationship, +1 indicates a perfect positive linear relationship, and 0 indicates no linear relationship.

## In our case there is no linear relationship.

R-squared (R-squared): This parameter measures the proportion of variance in the dependent variable that is explained by the independent variables. Higher R-squared values indicate a better fit to the data.

Adjusted R-squared (Adjusted_R_Squared): This parameter adjusts the R-squared value by the number of predictors and the sample size. It penalizes overfitting and provides a more conservative estimate of the model's fit.

F-statistic (F_Statistic) and its associated probability (Probability_F_Statistic): The F-statistic tests the overall significance of the model. A lower p-value indicates a more significant model fit.

Akaike Information Criterion (AIC) and Bayesian Information Criterion (BIC): These criteria provide a measure of the model's goodness of fit while penalizing models with more parameters. Lower AIC and BIC values indicate a better fit.

Standard errors of the coefficients (std_err_intercept, std_err_adjusted_year): These indicate the precision of the estimated coefficients. Smaller standard errors indicate more reliable estimates.

t-statistics (t_statistic_intercept, t_statistic_adjusted_year) and their associated p-values (p_value_intercept, p_value_adjusted_year): These measure the significance of individual coefficients. Lower p-values indicate a more significant relationship between the predictor and the response variable.

Confidence intervals (lower_bound_95_intercept, lower_bound_95_adjusted_year, upper_bound_95_intercept, upper_bound_95_adjusted_year): These provide a range of values within which the true coefficients are likely to fall. Narrower intervals indicate more precise estimates.

```python
# Connect to the database
conn = sqlite3.connect(FILE_PATH_TO_COMPILED_DATA + 'nba_data.db')

# Query the tables
query = "SELECT franchise_id, Team, Year, season_performance, draft_power FROM combined_season"
modeling_df = pd.read_sql_query(query, conn)

# Convert franchise_id column to float and then to int
modeling_df['franchise_id'] = modeling_df['franchise_id'].astype(float).astype(int)

# Sort the DataFrame by franchise_id and Year
modeling_df.sort_values(['franchise_id', 'Year'], inplace=True)

# Define the new columns
modeling_columns = ['franchise_id', 'Team', 'Year', 
                    'season_performance', 'sp_minus_1', 'sp_minus_2', 'sp_minus_3',
                    'draft_power', 'dp_minus_1', 'dp_minus_2', 'dp_minus_3']


def add_previous_seasons(group):
    shifted_values1 = pd.concat([group['season_performance'].shift(i, fill_value=np.nan) for i in range(1, 4)], axis=1)
    shifted_values1.columns = [f'sp_minus_{i}' for i in range(1, 4)]
    
    shifted_values2 = pd.concat([group['draft_power'].shift(i, fill_value=np.nan) for i in range(1, 4)], axis=1)
    shifted_values2.columns = [f'dp_minus_{i}' for i in range(1, 4)]
    
    group = pd.concat([group, shifted_values1, shifted_values2], axis=1)
    return group


# Apply the function to each group
modeling_df = modeling_df.groupby('franchise_id', group_keys=False).apply(add_previous_seasons)

# Reset the index
modeling_df.reset_index(drop=True, inplace=True)

modeling_df = modeling_df.assign(adjusted_sp=modeling_df['season_performance'] + modeling_df['sp_minus_1']**(3/5) + 
                                 modeling_df['sp_minus_2']**(1/2) + modeling_df['sp_minus_3']**(1/2))
modeling_df = modeling_df.assign(adjusted_dp=modeling_df['draft_power'] + modeling_df['dp_minus_1']**(3/5) + 
                                 modeling_df['dp_minus_2']**(1/2) + modeling_df['dp_minus_3']**(1/2))

nonzero_mask = modeling_df['adjusted_dp'] != 0
min_nonzero = np.min(modeling_df.loc[nonzero_mask, 'adjusted_dp'])
adjusted_dp = np.where(nonzero_mask, modeling_df['adjusted_dp'], min_nonzero)
modeling_df['log_adjusted_dp'] = np.log10(adjusted_dp)

modeling_df = modeling_df.assign(adjusted_year=modeling_df['Year'] - FIRST_YEAR)
modeling_df = modeling_df.assign(adjusted_year2=modeling_df['adjusted_year']**2)

modeling_columns.append('adjusted_sp')
modeling_columns.append('adjusted_dp')
modeling_columns.append('adjusted_year')
modeling_columns.append('adjusted_year2')

# Print the resulting DataFrame
# print(tabulate(modeling_df, headers=modeling_columns, tablefmt='psql', showindex=False))

# Close the connection
conn.close()

modeling_df.to_csv(FILE_PATH_TO_COMPILED_DATA+"lagged_team_performance.csv", index=False)
```

```python
def RunLinearModel_AdjustedSP(team_data):
    model = smf.ols('adjusted_sp ~ adjusted_year + log_adjusted_dp', data=team_data)
    results = model.fit()
    return model, results

def RunQuadraticModel_AdjustedSP(team_data):
    model = smf.ols('adjusted_sp ~ adjusted_year + adjusted_year2 + log_adjusted_dp', data=team_data)
    results = model.fit()
    return model, results

# Function to get the team name for a given franchise id
def getTeamName(franchise_id):
    team_name = team_mapping_df[team_mapping_df['franchise_id'] == franchise_id]['Team'].iloc[-1]
    return team_name

def getResultsFrame(results):
    results_data = [
        results.rsquared,
        results.rsquared_adj,
        results.fvalue,
        results.f_pvalue,
        results.nobs,
        results.aic,
        results.df_resid,
        results.bic,
        len(results.params),
        results.params[0],
        results.params[1],
        results.bse[0],
        results.bse[1],
        results.tvalues[0],
        results.tvalues[1],
        results.pvalues[0],
        results.pvalues[1],
        results.conf_int()[0][0] if len(results.conf_int()) > 0 else None,
        results.conf_int()[1][0] if len(results.conf_int()) > 0 else None,
        results.conf_int()[0][1] if len(results.conf_int()) > 0 else None,
        results.conf_int()[1][1] if len(results.conf_int()) > 0 else None
    ]

    results_summary_df = pd.DataFrame(
        [results_data],
        columns=[
            "R_squared",
            "Adjusted_R_squared",
            "F_statistic",
            "Probability_F_statistic",
            "Observations",
            "Akaike_Information_Criterion",
            "Degrees_of_Freedom_Residuals",
            "Bayesian_Information_Criterion",
            "Degrees_of_Freedom_Model",
            "coefficient_intercept",
            "coefficient_adjusted_year",
            "std_err_intercept",
            "std_err_adjusted_year",
            "t_statistic_intercept",
            "t_statistic_adjusted_year",
            "p_value_intercept",
            "p_value_adjusted_year",
            "lower_bound_95_intercept",
            "lower_bound_95_adjusted_year",
            "upper_bound_95_intercept",
            "upper_bound_95_adjusted_year"
        ]
    )
    return results_summary_df

franchise_columns = ['franchise_id', 'Team', 'lr_R_Squared', 'lr_F_Statistic', 'lr_Probability_F_Statistic', 
                     "lr_p_value_intercept", "lr_p_value_adjusted_year",
                     'qr_R_Squared', 'qr_F_Statistic', 'qr_Probability_F_Statistic', 
                     "qr_p_value_intercept", "qr_p_value_adjusted_year"]

franchise_df = pd.DataFrame(columns = franchise_columns)
print(modeling_df['franchise_id'])
franchise_range = range(1, 30)
for franchise in franchise_range:
    team_data = modeling_df[modeling_df['franchise_id'] == franchise].copy()
    if team_data.empty:
        print("Empty for franchise: "+str(franchise))
        continue

    # adjust for starting year.
    linear_model, linear_results = RunLinearModel_AdjustedSP(team_data)
    quadratic_model, quadratic_results = RunQuadraticModel_AdjustedSP(team_data)
    linear_results_df = getResultsFrame(linear_results)
    quadratic_results_df = getResultsFrame(quadratic_results)
    franchise_record_df = pd.DataFrame(columns=franchise_columns)
    franchise_record_df.loc[0, 'franchise_id'] = franchise
    franchise_record_df.loc[0, 'Team'] = getTeamName(franchise)
    franchise_record_df["lr_R_Squared"] = linear_results_df["R_squared"]
    franchise_record_df["lr_F_Statistic"] = linear_results_df["F_statistic"]
    franchise_record_df["lr_Probability_F_Statistic"] = linear_results_df["Probability_F_statistic"]
    franchise_record_df["lr_p_value_intercept"] = linear_results_df["p_value_intercept"]
    franchise_record_df["lr_p_value_adjusted_year"] = linear_results_df["p_value_adjusted_year"]
    franchise_record_df["qr_R_Squared"] = quadratic_results_df["R_squared"]
    franchise_record_df["qr_F_Statistic"] = quadratic_results_df["F_statistic"]
    franchise_record_df["qr_Probability_F_Statistic"] = quadratic_results_df["Probability_F_statistic"]
    franchise_record_df["qr_p_value_intercept"] = quadratic_results_df["p_value_intercept"]
    franchise_record_df["qr_p_value_adjusted_year"] = quadratic_results_df["p_value_adjusted_year"]
    franchise_df = pd.concat([franchise_df, franchise_record_df], ignore_index=True)


common_model = smf.ols('adjusted_sp ~ adjusted_year + log_adjusted_dp', data=modeling_df).fit()

print(common_model.summary())

common_model = smf.ols('adjusted_sp ~ log_adjusted_dp + adjusted_year2', data=modeling_df).fit()

print(common_model.summary())


franchise_df.to_csv(FILE_PATH_TO_COMPILED_DATA+"franchise_modeling.csv", index=False)
```

```python
def RunLinearModel(team_data):
    # Create lagged variable for draft_power
    team_data['lagged_draft_power'] = team_data['draft_power'].shift(1)  # Shift the values by one season
    
    # Remove the first row with missing lagged_draft_power
    team_data = team_data.dropna(subset=['lagged_draft_power'])
    
    # Define the regression formula with the lagged variable
    formula = 'season_performance ~ draft_power + lagged_draft_power'
    
    model = smf.ols(formula, data=team_data)
    results = model.fit()
    return model, results
```
